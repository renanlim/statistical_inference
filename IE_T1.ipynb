## 1 Teorema do Limite Central

Nesta parte, iremos apresentar uma simulação computacional envolvendo a distribuição exponencial e o Teorema do Limite Central.

Vamos definir alguns termos que vamos gerar n números aleatórios seguindo a distribuição expoencial utilizando a linguagem Python:

i) o lambda como parâmetro da distribuição expoencial;

ii) n indica a quantidade de valores a gerar;


```python
lambda_exp = 0.2
exps = [np.mean(np.random.exponential(scale=1/lambda_exp, size=40)) for i in range(1000)]

sns.histplot(data=exps, kde=True)
```

 iii) Calcule a média e variância aproximadas para a variável x e use o TLC para obter aproximações para a média e a variância da população subjacente.

```python
media_amostral = round(np.mean(exps), 3)
media_teórica = 1/lambda_exp

print("Média amostral:", media_amostral)
print("Média teórica:", media_teórica)
```

O Teorema do Limite Central afirma que, à medida que o tamanho da amostra aumenta, a distribuição amostral da média se aproxima cada vez mais de uma distribuição normal. Dessa forma, podemos observar que, à medida que o número de simulações aumenta, a média amostral e a média teórica se assemelham cada vez mais, corroborando o Teorema do Limite Central. Em outras palavras, à medida que a amostra se torna maior, a variabilidade entre as amostras é reduzida, resultando em uma distribuição cada vez mais próxima de uma normal. Isso reforça a importância de coletar amostras grandes o suficiente para fazer inferências precisas sobre a população subjacente.

# 2 Distribuições Amostrais

Nessa parte, iremos considerar uma população de sacos de batatas de 5Kg cada.

 i) Histograma da distribuição da população com relação à característica quantidade de batatas.

```python
batatas = np.arange(10, 21)
amostra_batatas = np.random.choice(batatas, size=10000)

sns.set()
sns.histplot(x=amostra_batatas, bins=10)
plt.xlabel('Quantidade de batatas')
plt.ylabel('Frequência')
plt.title('Distribuição da população de sacos de batatas')
```

Esse histograma mostra a distribuição da população de sacos de batatas, onde a quantidade de batatas em cada saco varia de 10 a 20. O eixo x representa a quantidade de batatas em cada saco, e o eixo y representa a frequência, ou seja, quantas vezes cada quantidade de batatas aparece na população. É possível observar que a distribuição é uniforme, ou seja, cada quantidade de batatas tem a mesma chance de aparecer na população.

ii) Média, variância e desvio padrão da população

```python
media = np.mean(amostra_batatas)
variancia = np.var(amostra_batatas)
desvio_padrao = np.sqrt(variancia)

print('População')
print("Média:", media)
print("Variância:", variancia)
print("Desvio padrão:", desvio_padrao)
```

iii) Distribuição amostral da média

```python
def distribuicao_amostral_media_n2():
    amostra = [np.random.choice(batatas, size=2) for i in range(100)]
    amostra_media = [np.mean(amostra[i]) for i in range(len(amostra))]
    
    return amostra_media
```

```python
media_amostra = distribuicao_amostral_media_n2()
plt.xlabel("Casos")
plt.ylabel("Amostras de tamanho 2")
plt.axhline(np.mean(media_amostra), color='r', linestyle='dashed', linewidth=2)
plt.plot(media_amostra)
```

A linha vermelha horizontal representa a média da distribuição amostral de média. O gráfico mostra que a maioria das médias amostrais está próxima da média da população que é 15, e que a distribuição amostral de média é aproximadamente normal.

Note que o gráfico é uma boa representação visual do Teorema do Limite Central, já que mostra que a distribuição amostral de média se aproxima de uma distribuição normal à medida que o tamanho da amostra aumenta.

```python
plt.xlabel("Média das combinações")
plt.ylabel("Densidade")
plt.title("Distribuição amostral da média")
plt.xticks(batatas)
sns.distplot(media_amostra)
```

No gráfico em questão, é possível ver que a distribuição amostral da média para n = 2 é aproximadamente normal, o que está de acordo com o Teorema do Limite Central. Além disso, a média da distribuição amostral é de aproximadamente 15.

vi) Estatística da amostra para estimar a médida da população

Para estimar a média da população a partir de uma amostra de tamanho 2 tomada aleatoriamente, calcularia a média amostral da amostra, que é a soma das observações dividida pelo tamanho da amostra.

Embora a média amostral seja um estimador comumente utilizado para estimar a média da população, não há garantia de que a média amostral seja um bom estimador da população, pois o valor da média amostral pode variar consideravelmente de amostra para amostra.

Uma outra estatística que pode ser um melhor estimador seria a mediana da amostra. A mediana é uma medida de tendência central que é menos sensível a valores extremos do que a média. No entanto, a escolha da estatística dependerá do objetivo e da natureza dos dados em questão. É sempre importante considerar a distribuição dos dados e a possibilidade de valores extremos ou discrepantes na amostra.

vii) Distribuição amostral da média para amostras de tamanho n = 9

a) 

O número de amostras possíveis de tamanho n = 9 é dado por combinação de 20 elementos tomados 9 a 9, ou seja:

C(20,9) = 20! / (9! * (20-9)!) = 167,960

Gerar todas essas amostras manualmente seria inviável, pois a quantidade de amostras possíveis é muito grande. Mesmo usando um computador, gerar todas as amostras seria um desafio, já que demandaria muito processamento e memória. Por isso, geralmente usamos métodos estatísticos e amostragem para estimar a distribuição amostral, ao invés de gerá-la teoricamente.

b)
 

```python
def distribuicao_amostral_media_n9():
  
    pop = np.arange(10, 21)
    amostras = np.random.choice(pop, size=(100000, 9), replace=True)
    medias = np.mean(amostras, axis=1)

    return medias

dist_amostral = distribuicao_amostral_media_n9()
sns.set()
sns.histplot(x=dist_amostral, bins=20)
```

Note que estamos gerando 100000 amostras aleatórias de tamanho 9 com substituição, o que deve ser suficiente para obter uma boa aproximação da distribuição amostral empírica.

# 3  Intervalo de confiança para µ

Nessa etapa será realizada a importação de dados do arquivo SchroederEpley2015data.txt

```python
dataset_url = "SchroederEpley2015data.txt"

df_dataframe = pd.read_csv(dataset_url)
df_dataframe.head()
```

Valores importantes:
 - Coluna: CONDITION, o valor 1 indica que o recrutador foi alocado na condição de audio; 
 - Coluna: CONDITION, o valor 0 indica que o recrutador foi alocado na condiçãoo de transcrição;
 - Coluna: Intellect_Rating, indica a avaliação que cada recrutador atribui ao candidato (inteiro de 0 a 10);
 - Sa são amostras que correspodem aos valores de avalição na condição de aúdio;
 - St são amostras que correspodem aos valores de avalição na condição de transcrição;


i) Compute a média e o tamanho (quantidade de observações) tanto para Sa quanto St.

```python
# Separando as amostras Sa e St
sa = df_dataframe[df_dataframe['CONDITION'] == 1]['Intellect_Rating']
st = df_dataframe[df_dataframe['CONDITION'] == 0]['Intellect_Rating']

# Calculando a média e o tamanho para Sa
media_sa = sa.mean()
tamanho_sa = sa.shape[0]

# Calculando a média e o tamanho para St
media_st = st.mean()
tamanho_st = st.shape[0]

print("Média Sa:", media_sa, "| Tamanho Sa:", tamanho_sa)
print("Média St:", media_st, "| Tamanho St:", tamanho_st)
```

ii) Boxplot

```python
# Criar o boxplot
plt.boxplot([sa, st])

# Adicionar títulos e rótulos
plt.title('Boxplot para Sa e St')
plt.xticks([1, 2], ['Condição de Áudio', 'Condição de Transcrição'])
plt.ylabel('Avaliação')
# Exibir o gráfico
plt.show()
```

iii) Teste de normalidade

```python
from scipy.stats import shapiro

# Teste de normalidade para Sa
stat_sa, p_sa = shapiro(sa)
print("Estatística de teste para Sa: ", stat_sa)
print("p-value para Sa: ", p_sa)

# Teste de normalidade para St
stat_st, p_st = shapiro(st)
print("Estatística de teste para St: ", stat_st)
print("p-value para St: ", p_st)
```

Se o valor de p do teste for menor do que um nível de significância pré-definido (geralmente 0,05), podemos rejeitar a hipótese nula e concluir que os dados não seguem uma distribuição normal. Assim, 
valores de p são maiores do que 0,05, indicando que não há evidência suficiente para rejeitar a hipótese nula de que as amostras seguem uma distribuição normal. Portanto, podemos considerar que as amostras Sa e St seguem uma distribuição normal.

iv) Intervalo de confiança para a média de avaliações produzidas por duas populações distintas

```python
import scipy.stats as stats

media_sa = np.mean(sa)
media_st = np.mean(st)

desvio_padrao_sa = np.std(sa, ddof=1)
desvio_padrao_st = np.std(st, ddof=1)

tamanho_sa = len(sa)
tamanho_st = len(st)
```

```python
# Intervalo de confiança para Sa
intervalo_sa = stats.t.interval(0.95, tamanho_sa - 1, loc=media_sa, scale=desvio_padrao_sa / np.sqrt(tamanho_sa))

# Intervalo de confiança para St
intervalo_st = stats.t.interval(0.95, tamanho_st - 1, loc=media_st, scale=desvio_padrao_st / np.sqrt(tamanho_st))

print(f"Intervalo de confiança para Sa: [{intervalo_sa[0]:.2f}, {intervalo_sa[1]:.2f}]")
print(f"Intervalo de confiança para St: [{intervalo_st[0]:.2f}, {intervalo_st[1]:.2f}]")
```

A interpretação dos intervalos de confiança para Sa e St é que, com 95% de confiança, a média de avaliações produzidas pelos recrutadores que ouvem os áudios está entre 4.90 e 6.37, e a média de avaliações produzidas pelos recrutadores que leem os transcritos está entre 2.70 e 4.60. Issi significa que há diferenças significativas nas avaliações feitas pelos recrutadores que ouvem os áudios do que os que leem os transcritos.

v) A conclusão de Schoroeder e Epley é válida?

Com base nos intervalos de confiança obtidos no item anterior, podemos observar que a média de avaliações produzidas pelos recrutadores que ouvem os áudios (Sa) é significativamente maior do que a média de avaliações produzidas pelos recrutadores que leem os transcritos (St). Além disso, os intervalos de confiança para as duas amostras não se sobrepõem, o que indica uma diferença significativa entre as duas médias.

Portanto, com base nos resultados encontrados, a conclusão a que chegaram os autores em Schroeder and Epley (2015) parece ser válida, ou seja, que os recrutadores tendem a avaliar melhor os candidatos quando os ouvem em vez de lê-los.

## 4 Intervalo de confiança para µ

Nessa etapa será realizada a importação de dados do arquivo survey.csv

```python
df_survey = pd.read_csv('survey.csv')
df_survey.head ()
```

Valores importantes:
- Sex. O sexo do aluno. (Fator com os níveis Male e Female.);
- Wr.Hnd. vão (distância da ponta do polegar à ponta do dedo mínimo da mão aberta) da ão que escreve, em centímetros;
- NW.Hnd. extensão da mão que não escreve;
-  W.Hnd. mão de escrita. (Left ou Right.);
- Fold. Cruze os braços! Qual estão no topo? (R on L, L on R, Neither.);
- Pulse. taxa de pulso do aluno (batimentos por minuto);
- Clap. Bata palmas! Qual mão está para cima? (Right, Left,None.);
- Exer. quantas vezes o aluno se exercita. (Freq (frequentemente), Some, None.);
- Smoke. Quanto o aluno fuma. (Heavy, Regul (regularmente), Occas (occasionalmente), Never.);
- Height. altura do aluno em centímetros;
- M.I. se o aluno expressou a altura em unidades imperiais (pés/polegadas) ou métricas (centímetros/metros). (Metric, Imperial.);
- Age. Idade do aluno em anos;

i) Calculo do intervalo de confiança no nível de 95% para a altura média dos estudantes da universidade.

```python
df_survey.dropna(subset=['Height'], inplace=True)
```

Isso removerá todas as linhas em que a coluna "Height" contém valores faltantes.

```python
# Extrair as alturas da coluna Height
heights = df_survey['Height']

# Calcular a média e o desvio padrão amostral das alturas
mean = np.mean(heights)
std = np.std(heights, ddof=1)

# Calcular o tamanho da amostra
n = len(heights)

# Calcular o intervalo de confiança de 95% usando a distribuição t de Student
t = stats.t.ppf(0.975, df=n-1)
ic_inferior = mean - t * std / np.sqrt(n)
ic_superior = mean + t * std / np.sqrt(n)

# Imprimir o resultado
print("Intervalo de confiança de 95% para a altura média dos estudantes: [{:.2f}, {:.2f}]".format(ic_inferior, ic_superior))
```

ii) Intervalo de confiança usando o z-score.

```python
altura = df_survey['Height']

# Calculando a média e o desvio padrão da amostra
media = altura.mean()
desvio_padrao = altura.std()

print("Média da amostra:", media)
print("Desvio padrão da amostra:", desvio_padrao)
```

```python
from scipy.stats import norm

# Definindo o nível de confiança e o valor crítico do z-score
confianca = 0.95
z = norm.ppf((1 + confianca) / 2)

# Calculando o intervalo de confiança
ic_inferior = media - z * (desvio_padrao / np.sqrt(len(altura)))
ic_superior = media + z * (desvio_padrao / np.sqrt(len(altura)))

print("Intervalo de confiança usando z-score:", "[",ic_inferior, ic_superior,"]")
```

Portanto, podemos concluir que, com 95% de confiança, a altura média dos estudantes da universidade está entre 171.0 cm e 173.7 cm.

iii) Análise comparativa dos dois intervalos de confiança obtidos

Os intervalos de confiança obtidos usando o t-score e o z-score são muito próximos. Isso ocorre porque a amostra é grande, o que faz com que a distribuição t de Student se aproxime cada vez mais da distribuição normal padrão quanto maior o tamanho da amostra.

No entanto, é importante destacar que o intervalo de confiança usando o t-score é ligeiramente mais amplo do que o intervalo de confiança usando o z-score. Isso ocorre porque a distribuição t de Student é um pouco mais "achatada" nas caudas do que a distribuição normal padrão. Portanto, para um mesmo nível de confiança, é necessário usar um valor maior de t do que de z para a obtenção do intervalo de confiança.

## 5 Intervalo de confiança para p

```python
df_survey = pd.read_csv('survey.csv')

# Contando o número de destros e canhotos
contagem = df_survey['W.Hnd'].value_counts()
print(contagem)
```

Se a saída mostrar que tanto "Right" quanto "Left" possuem contagens maiores ou iguais a 10, então podemos prosseguir com a construção do intervalo de confiança. Caso contrário, será necessário aumentar o tamanho da amostra ou utilizar uma amostra diferente que atenda aos requisitos.







i) Intervalo de confiança, no nível de 90%, para a proporção de alunos da universidade que são canhotos

```python
# Eliminando os valores faltantes na coluna W.Hnd
df = df_survey.dropna(subset=['W.Hnd'])

# Contando o número de alunos canhotos
num_canhotos = (df['W.Hnd'] == 'Left').sum()

# Calculando a proporção amostral de alunos canhotos
p = num_canhotos / df.shape[0]

# Calculando o desvio padrão da proporção amostral
s = np.sqrt(p * (1 - p) / df.shape[0])

# Definindo o nível de confiança
conf_level = 0.9

# Calculando o valor crítico da distribuição normal padrão
z = norm.ppf(1 - (1 - conf_level)/2)

# Calculando o erro padrão da proporção amostral
se = z * s

# Calculando o intervalo de confiança
inferior = p - se
superior = p + se

print("Intervalo de confiança para a proporção de alunos canhotos:")
print("({:.2%}, {:.2%})".format(inferior, superior))
```

ii) Análise do problema

O intervalo de confiança obtido indica que, com um nível de confiança de 90%, a proporção de alunos canhotos na universidade está entre 4,79% e 10,47%. Isso significa que se tomarmos várias amostras da população, a proporção de canhotos estará dentro desse intervalo em aproximadamente 90% das vezes. É importante notar que o intervalo não inclui o valor de 10%, o que sugere que a proporção de canhotos na população pode ser menor do que se imaginava anteriormente. No entanto, é preciso ter em mente que a amostra pode não ser representativa da população em geral e que outras amostras podem levar a intervalos de confiança diferentes.
